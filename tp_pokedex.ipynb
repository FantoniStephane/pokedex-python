## Contexte

Une entreprise organise un tournoi Pokémon fictif pour animer une journée de cohésion d'équipe.

Avant le tournoi, les participants doivent créer une petite équipe de Pokémon. L'organisation souhaite disposer d'un outil en ligne de commande permettant de rechercher des Pokémon, d'afficher leurs caractéristiques principales, puis de construire une mini-équipe.

Tu dois développer ce programme en Python en utilisant une API publique : **PokéAPI**.

---

## Objectif du TP

Développer un programme Python permettant de créer une mini-équipe Pokémon à partir de données récupérées en ligne.

Le programme doit permettre de :
1. Rechercher un Pokémon par son nom
2. Récupérer ses informations depuis une API
3. Afficher une fiche lisible du Pokémon
4. Proposer d'ajouter ou non ce Pokémon à une équipe
5. Afficher l'équipe finale

---

## API à utiliser

```
https://pokeapi.co/api/v2/pokemon/{nom}
```

Exemple :
```
https://pokeapi.co/api/v2/pokemon/pikachu
```

L'endpoint permet de récupérer les informations d'un Pokémon à partir de son nom ou de son identifiant.

---

## Données à exploiter

Pour chaque Pokémon recherché, le programme devra afficher au minimum :
- Son **nom**
- Sa **taille**
- Son **poids**
- Son ou ses **types**
- Au moins **3 statistiques de base** : points de vie, attaque, défense
